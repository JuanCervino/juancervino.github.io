```python
import yaml
import io


def write_author_line(authors : list, collaborators : dict):
    line = ""
    for count, author in enumerate(authors):
        if "web" not in collaborators[author]:
            line = line + str(collaborators[author]['name'] )
        else:
            line = line + f" <a href='{str(collaborators[author]['web'])}' target='_blank'>{str(collaborators[author]['name'] )}</a>" 

        if count < len(authors) - 1:
            line = line + ", "
        
    return line+'<br>'

with open("yamls/collaborators.yaml", 'r') as pubs:
    collaborators = yaml.safe_load(pubs)

# Read YAML file
with open("yamls/pubs.yaml", 'r') as pubs:
    pubs_data = yaml.safe_load(pubs)


with open("index_middle.html", 'w') as pub_file:

    for key in pubs_data.keys():
        pub_file.write(f"	<h3>{key}</h3> \n")
        for item in pubs_data[key]:
            pub_file.write('<div class="infoblock"><div class="blockcontent"><p>\n')
            pub_file.write(f"<b>{item['title']}</b><br>\n")
            pub_file.write(write_author_line(item['authors'], collaborators))
            pub_file.write(f"{item['venue']}<a href='{item['paper_pdf']}' target='_blank'>[PDF]</a>")
            if 'code' in item:
                pub_file.write(f" <a href='{item['code']}' target='_blank'>[Code]</a>")

                
            pub_file.write('</p></div></div><br>\n')

def merge_html_files(file_paths, output_file):
    merged_content = ""

    # Read contents of each HTML file and merge
    for file_path in file_paths:
        with open(file_path, "r") as file:
            merged_content += file.read()

    # Write merged content to the output file
    with open(output_file, "w") as file:
        file.write(merged_content)

# Example usage:
file_paths = ["index_pre.html", "index_middle.html", "index_post.html"]
output_file = "index.html"

merge_html_files(file_paths, output_file)
```
